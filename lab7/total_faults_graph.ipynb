```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the provided data
data = {
    'Cache Size': [10, 50, 100, 250, 500],
    'FIFO': [9916, 9515, 9018, 7534, 5130],
    'LRU': [9915, 9510, 9029, 7532, 5206],
    'Second Chance': [9915, 9510, 9022, 7526, 5178]
}

df = pd.DataFrame(data)

# Melt the DataFrame to a long format suitable for seaborn
df_melted = df.melt('Cache Size', var_name='Algorithm', value_name='Page Faults')

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_melted, x='Cache Size', y='Page Faults', hue='Algorithm', marker='o')

# Customize the plot
plt.title('Page Faults for Different Page Replacement Algorithms')
plt.xlabel('Cache Size')
plt.ylabel('Page Faults')
plt.legend(title='Algorithm')
plt.grid(True)

# Show the plot
plt.show()
```
